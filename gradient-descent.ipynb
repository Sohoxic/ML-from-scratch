```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample data (y = 2x + 3 with some noise)
np.random.seed(0)
X = np.linspace(0, 10, 50)
noise = np.random.randn(50)

print(X)
print(f"Shape: {X.shape}") 

print(f"Shape: {noise.shape}")
print(f"Sample: {noise[:5]}")  

y = 2*X + 3 + np.random.randn(50)

# Parameters
m, b = 0.0, 0.0   # initial guess
lr = 0.01         # learning rate
epochs = 1000     # iterations 
```

```python

n = len(X)

for epoch in range(epochs):
    y_pred = m*X + b
    error = y - y_pred
    
    # gradients
    dm = -(2/n) * np.dot(X, error)
    db = -(2/n) * np.sum(error)
    
    # update step
    m -= lr * dm
    b -= lr * db
    
    if epoch % 100 == 0:
        loss = np.mean(error**2)
        print(f"Epoch {epoch}: Loss={loss:.4f}, m={m:.4f}, b={b:.4f}")

# Final results
print(f"\nFinal parameters: m={m:.2f}, b={b:.2f}")

# Plot fit
plt.scatter(X, y, label="Data")
plt.plot(X, m*X + b, color="red", label="Fitted line")
plt.legend()
plt.show()
```
